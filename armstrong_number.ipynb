```python
# armstrong_number.py
num = int(input("Enter a number: "))
temp = num
sum_of_powers = 0
digits = len(str(num))

while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** digits
    temp //= 10

if num == sum_of_powers:
    print(f"{num} is an Armstrong Number ✅")
else:
    print(f"{num} is not an Armstrong Number ❌")
```
